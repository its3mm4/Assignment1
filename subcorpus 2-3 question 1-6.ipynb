```python
# Corpus Two
corpus2 = "/Users/tammychen/Downloads/Ling 250/TheHoundoftheBaskervilles.txt"
corpus2File = open(corpus2).read()
corpus2File
```

```python
# question 1

text = corpus2File
words = text.split()
tokens = len(words)

print(f"The length of the text in words is: {tokens}")
```

```python
# question 2

types = len(set(words))

lexical_diversity = types / tokens

print(f"Lexical diversity: {lexical_diversity:}")
```

```python
# question 3

import nltk
from nltk import FreqDist
from nltk.corpus import stopwords
from string import punctuation

# Tokenize the text into words
words = nltk.word_tokenize(corpus2File)

# Remove stopwords and punctuation
stop_words = set(stopwords.words("english"))
filtered_words = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]

# Create a frequency distribution of the words
fdist1 = FreqDist(filtered_words)

# To get the most common words (e.g., the top 10)
common_words = fdist1.most_common(10)  # Change 10 to the desired number of top words
for word, freq in common_words:
    print(f"{word}: {freq}")
```

```python
# question 4

import re

# Sample text
text = corpus2File

# Use regular expressions to find words that are at least 10 characters long
long_words = re.findall(r'\b\w{10,}\b', text)

# Create a dictionary to count the occurrences of each long word
word_count = {}
for word in long_words:
    word_count[word] = word_count.get(word, 0) + 1

# Print the long words and their counts
for word, count in word_count.items():
    print(f"'{word}': {count}")

# If you want to sort the words by count in descending order:
sorted_word_count = sorted(word_count.items(), key=lambda x: x[1], reverse=True)
for word, count in sorted_word_count:
    print(f"'{word}': {count}")
```

```python
# question 5

import nltk
from nltk.corpus import gutenberg
from nltk.tokenize import sent_tokenize, word_tokenize

# Download the Gutenberg corpus if not already downloaded
nltk.download('gutenberg')

# Choose a specific Gutenberg text (replace 'shakespeare-hamlet.txt' with the desired text)
corpus2 = "/Users/tammychen/Downloads/Ling 250/TheHoundoftheBaskervilles.txt"
text_name = corpus2
text = gutenberg.raw(text_name)

# Tokenize the text into sentences
sentences = sent_tokenize(text)

# Find the longest sentence
longest_sentence = max(sentences, key=lambda sentence: len(word_tokenize(sentence)))

# Count the number of words in the longest sentence
words_in_longest_sentence = len(word_tokenize(longest_sentence))

# Print the longest sentence and the number of words
print(f"The longest sentence is: {longest_sentence}")
print(f"Number of words in the longest sentence: {words_in_longest_sentence}")
```

```python
# qusetion 6
from nltk.stem import PorterStemmer
nltk.download('punkt')

ps = PorterStemmer()
stemmed_longest_sentence = ' '.join([ps.stem(word) for word in word_tokenize(longest_sentence)])

# Print the stemmed version of the longest sentence
print(f"\nStemmed version of the longest sentence: {stemmed_longest_sentence}")
```

```python
# Corpus three

import requests
from bs4 import BeautifulSoup

# List of URLs to extract plain text from
urls = [
    "https://www.cbc.ca/news/world/gay-dating-app-thrives-in-china-where-lgbt-rights-are-lagging-1.4733218",
    "https://www.cbc.ca/news/canada/montreal/first-person-trans-rights-coming-out-parents-family-1.6979523",
    "https://www.cbc.ca/news/canada/british-columbia/taiwanese-in-canada-travel-to-island-to-elect-president-1.5412942",
    "https://capp-usa.org/male-female-transgenderism/"
]

# Open a file named 'news.txt' in write mode
with open('news.txt', 'w', encoding='utf-8') as file:
    for url in urls:
        response = requests.get(url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            if "cbc.ca" in url:
                main_content = soup.find('div', class_='story')
            elif "capp-usa.org" in url:
                main_content = soup.find('div', class_='entry-content')
            
            if main_content:
                # Extract the plain text
                plain_text = main_content.get_text()

                # Write the plain text to the file
                file.write(f"Plain text from {url}:\n{plain_text}\n\n")
            
```
